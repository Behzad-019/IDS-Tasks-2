```
pakistan_map = folium.Map(location = [30.37,69.34],zoom_start=4.5,tiles='OpenStreetMap')
```

```
from google.colab import drive
drive.mount('/content/gdrive')
```

```
#df = pd.read_csv('gdrive/My Drive/Semester6/IDSTask2.csv')

df = pd.DataFrame(pd.read_excel('gdrive/My Drive/Semester6/IDSTask2.xlsx'))
df
```

```
t = 0
for c in df['Total Cases']:
  t = float(c)+float(t)
```

```
for city,lat,long,tc in zip(list(df['Cities']),list(df['Latitude']),list(df['Longitude']),list(df['Total Cases'])):
    folium.CircleMarker(location = [lat,long],
                       radius = (tc/t)*100,
                       color = 'red',
                       fill = False,
                       fill_color = 'blue').add_to(pakistan_map)
    xyz = '<strong>State:  ' + city + '</strong><br>' + 'Total Cases: ' + str(tc) + '<br>' 
    iframe = folium.IFrame(xyz, width=200, height=150)
    popupData = folium.Popup(iframe, max_width=500)
    folium.Marker(location = [lat,long], popup=popupData).add_to(pakistan_map)
    

pakistan_map
```
